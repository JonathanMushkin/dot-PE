```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(low=0, high=1, size=(2, 10000))

x = np.sort(x, axis=0)
```

```python
counts, bins, *_ = plt.hist(
    x[0], bins=np.linspace(0, 1, 100, endpoint=True), histtype="step"
)
counts, bins, *_ = plt.hist(x[1], bins=bins, histtype="step")
```
